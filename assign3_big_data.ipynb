```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1=pd.read_csv('./data.csv')
data1
```

```python
plt.plot(data1["month_number"],data1['total_profit'])
plt.plot("line plot")
plt.xlabel("month_number")
plt.ylabel("total profit")
plt.show()
```

```python
plt.plot(data1["month_number"],data1['total_profit'])
plt.plot("line plot")
plt.ylabel("month_number")
plt.xlabel("total profit")
plt.show()
```

```python
plt.plot(data1["month_number"],data1['facecream'],label=['facecream'])
plt.plot(data1["month_number"],data1['facewash'],label=['facewash'])
plt.plot(data1["month_number"],data1['shampoo'],label=['shampoo'])
plt.plot(data1["month_number"],data1['bathingsoap'],label=['bathingsoap'])
plt.plot(data1["month_number"],data1['moisturizer'],label=['moisturizer'])
plt.legend()
plt.plot("multiline plot")
plt.ylabel("month_number")
plt.xlabel("total profit")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming data1 is a DataFrame containing your data

# Plotting bars for each attribute
plt.bar(data1["month_number"], data1['facecream'], label='Face Cream')
plt.bar(data1["month_number"], data1['facewash'], label='Face Wash')
plt.bar(data1["month_number"], data1['shampoo'], label='Shampoo')
plt.bar(data1["month_number"], data1['bathingsoap'], label='Bathing Soap')
plt.bar(data1["month_number"], data1['moisturizer'], label='Moisturizer')

plt.legend()  # Show legend based on label
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("Sales by Product Over Months")
plt.show()
```
